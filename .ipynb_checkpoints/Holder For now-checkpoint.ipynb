### 4.2 Function basic pre-processing


```python
def preprocess_text(text):
    # 4.2.1. Lowercasing
    text = text.lower()

    # 4.2.2. Remove URLs and User(@)
    text = re.sub(r'http\S+|www\S+|https\S+|@[^\s]+', '', text, flags=re.MULTILINE)

    # 4.2.3. Remove Punctuation & Special Characters
    text = re.sub(r'[^\w\s]', '', text)

    # 4.2.4. Remove numeric numbers
    text = re.sub(r'\d+', '', text)

    # 4.2.5. Remove Special Characters & Handle Encoding
    text = ''.join(char for char in unicodedata.normalize('NFKD', text) if not unicodedata.combining(char))

    # 4.2.6. Remove extra white space & leading space
    text = ' '.join(text.split())

    # 4.2.7. Remove Stopwords
    stop_words = set(stopwords.words('english'))
    text = ' '.join(word for word in text.split() if word not in stop_words)

    return text
```

#### 4.3.1. Handle Hashtags(#)

- This code facilitates an exploration of dominant hashtags across various sentiment classes, offering a glimpse into prevalent themes and trending topics associated with each sentiment category in the `training dataset`.

```python
def hashtag_extract(tweet):

    hashtags = []

    for i in tweet:
        ht = re.findall(r"#(\w+)", i)
        hashtags.append(ht)

    hashtags = sum(hashtags, [])
    frequency = nltk.FreqDist(hashtags)

    hashtag_df = pd.DataFrame({'hashtag': list(frequency.keys()),
                       'count': list(frequency.values())})
    hashtag_df = hashtag_df.nlargest(15, columns="count")

    return hashtag_df

# You might need to replace 'message' with the actual column name in your DataFrame
df_temp_Hashtag = df_train.copy()
df_temp_Hashtag['message'] = df_temp_Hashtag['message'].apply(lambda x: x.lower())

# Extracting the hashtags from tweets in each class
pro_hashtag = hashtag_extract(df_temp_Hashtag['message'][df_temp_Hashtag['sentiment'] == 1])
anti_hashtag = hashtag_extract(df_temp_Hashtag['message'][df_temp_Hashtag['sentiment'] == -1])
neutral_hashtag = hashtag_extract(df_temp_Hashtag['message'][df_temp_Hashtag['sentiment'] == 0])
news_hashtag = hashtag_extract(df_temp_Hashtag['message'][df_temp_Hashtag['sentiment'] == 2])
```

- We have all the hashtags stored in data frames `pro_hashtag`, `anti_hashtag`, `neutral_hashtag` & `new_hashtag`

### 4.4 Advanced Text Processing Techniques

- **Advanced Text Processing Techniques** involve sophisticated methods to enhance the structure and understandability of textual data.
- The goal is to create a more refined and standardized representation of text, aiding in the effectiveness of subsequent analysis and machine learning tasks.

#### 4.4.1. Tokenization
- **Tokenization** is the process of breaking down a text into smaller units, typically words or subwords, to facilitate analysis and natural language processing tasks.

```python
def tokenizer(text):

    # Tokenization
    tokens = word_tokenize(text)
    return tokens
```

```python
# Apply tokenization to the 'message' column in df_train
df_combined['tokens'] = df_combined['clean_message'].apply(tokenizer)

df_combined[['clean_message', 'tokens', 'sentiment']].head()
```

#### 4.4.2 Lemmatization
- **Lemmatization** is a process that involves reducing words to their base or root form, known as a lemma, to achieve normalization and enhance the consistency of textual analysis to aid in standardizing and simplifying the text for improved analysis and machine learning model training.

```python
# Function to perform Lemmatization on a sentence
def lemmatizer(text):

    # Tokenization
    tokens = word_tokenize(text)

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    return tokens
```

```python
# Apply Lemmatization to the 'clean_message' column in df_train
df_combined['lemmatized'] = df_combined['clean_message'].apply(lemmatizer)

# Reorder columns to have 'lemmatized' next to 'tokens'a
df_combined = df_combined[['tweetid','message','clean_message', 'tokens', 'lemmatized','sentiment']]
```

### 5.1 Climate change trending words

```python
def frequency_by_sentiment(tweet_series, sentiment_value):

    # Filter tweets for the specified sentiment class
    filtered_tweets = tweet_series[df_train['sentiment'] == sentiment_value]

    # Join the list of stemmed words into a space-separated string
    filtered_tweets_str = filtered_tweets.apply(lambda x: ' '.join(x))

    # Count vectorizer excluding English stopwords
    cv = CountVectorizer(stop_words='english')
    words = cv.fit_transform(filtered_tweets_str)

    # Count the words in the tweets and determine the frequency of each word
    sum_words = words.sum(axis=0)
    words_freq = [(word, sum_words[0, i]) for word, i in cv.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)

    # Create a dataframe to store the top 25 words and their frequencies
    frequency = pd.DataFrame(words_freq, columns=['word', 'freq'])
    frequency = frequency.head(25)

    return frequency
```

```python
# Extract the top 25 words in each class
pro_frequency = frequency_by_sentiment(df_train['lemmatized'],1)
anti_frequency = frequency_by_sentiment(df_train['lemmatized'],-1)
news_frequency = frequency_by_sentiment(df_train['lemmatized'],2)
neutral_frequency = frequency_by_sentiment(df_train['lemmatized'],0)
```

```python
# Extract the words in the tweets for the pro and anti climate change classes
anti_words = ' '.join([text for text in anti_frequency['word']])
pro_words = ' '.join([text for text in pro_frequency['word']])
news_words = ' '.join([text for text in news_frequency['word']])
neutral_words = ' '.join([text for text in neutral_frequency['word']])

# Create word clouds with adjusted parameters
anti_wordcloud = WordCloud(width=800, height=500, max_font_size=110, background_color='white', colormap="Reds", random_state=73).generate(anti_words)
pro_wordcloud = WordCloud(width=800, height=500, max_font_size=110, background_color='white', colormap="Greens", random_state=23).generate(pro_words)
news_wordcloud = WordCloud(width=800, height=500, max_font_size=110, background_color='white', colormap="Blues", random_state=11).generate(news_words)
neutral_wordcloud = WordCloud(width=800, height=500, max_font_size=110, background_color='white', colormap="Greys", random_state=11).generate(neutral_words)
```

### 5.3 Modeling Preparation

- Model preparation in the context of machine learning refers to the set of activities and steps taken to get a model ready for training and evaluation.
- Model preparation aims to create an environment in which the machine learning model can learn from the data effectively, make accurate predictions, and generalize well to new, unseen examples.


#### 5.3.1 Dataset Split

- The process of data splitting involves dividing the dataset into two subsets: one for `training` the machine learning model and another for `testing` its performance.
- The data will be split **80%** for `training` and teh remaining **20%** for `testing`.


```python
X = df_train['tokens'].apply(lambda tokens: " ".join(tokens))
Y = df_train['sentiment']

# Split the data into training and validation sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=GLOBAL_CONSTANT)
```

#### 5.3.2 Handling Imbalanced Classes - SMOTE

- **SMOTE** (Synthetic Minority Over-sampling Technique) is a method used to address the imbalance in your dataset by creating synthetic examples of the underrepresented `sentiment` classes.
- Essentially, it helps balance the number of tweets across different `sentiment` categories (like *Pro*, *Anti*, *Neutral*, *News*) by generating new, synthetic tweets for the minority classes, improving the model's ability to learn from a more evenly distributed dataset.

##### Text Vectorization(Pre-SMOTE)

```python
# Create CountVectorizer for basic numerical encoding
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train)
```

```python
# Create SMOTE instance for resampling
smote = SMOTE(random_state=42)
# Resampling the training data using SMOTE
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_counts, y_train)
```

#### 5.3.3 Vectorization -Term Frequency-Inverse Document Frequency (TF-IDF)

- **TF-IDF** is a numerical representation used in natural language processing to gauge the significance of words in individual documents.
- **TF-IDF** assigns higher weights to words that are both frequently used within specific tweets and relatively rare across the entire dataset.
-  This approach captures the distinctive language patterns within tweets, aiding in the accurate representation of textual data for sentiment analysis.


```python
# Transform count vectorized data back to text
X_train_resampled_text = count_vect.inverse_transform(X_train_resampled)

# Creating a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5500, ngram_range=(1,2))
X_train_tfidf = tfidf_vectorizer.fit_transform([' '.join(doc) for doc in X_train_resampled_text])
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

##### Class distribution - Before & After SMOTE

```python
# Define the colors for each class
colors = {-1: 'lightcoral', 0: 'lightgrey', 1: 'lightgreen', 2: 'lightskyblue'}

# Count the occurrences of each class
class_counts = y_train.value_counts().sort_index()

# Creating a bar plot for the class distribution
plt.figure(figsize=(10, 6))
for class_label in class_counts.index:
    plt.bar(class_label, class_counts[class_label], color=colors[class_label], edgecolor='black')

plt.title("Class Distribution Before SMOTE")
plt.xlabel("Sentiment Classes")
plt.ylabel("Count")
plt.xticks(np.unique(y_train))  # Set x-ticks to be the unique classes
plt.show()
```

```python
# Define the colors for each class
colors = {-1: 'lightcoral', 0: 'lightgrey', 1: 'lightgreen', 2: 'lightskyblue'}

# Convert y_train_resampled to a Pandas Series and count the occurrences of each class
y_train_resampled_series = pd.Series(y_train_resampled)
class_counts_resampled = y_train_resampled_series.value_counts().sort_index()

# Creating a bar plot for the class distribution after SMOTE
plt.figure(figsize=(10, 6))
for class_label in class_counts_resampled.index:
    plt.bar(class_label, class_counts_resampled[class_label], color=colors[class_label], edgecolor='black')

plt.title("Class Distribution After SMOTE")
plt.xlabel("Sentiment Classes")
plt.ylabel("Count")
plt.xticks(class_counts_resampled.index)  # Set x-ticks to be the unique classes
plt.show()
```

```python
# Define the classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=GLOBAL_CONSTANT)

# Create the imbalanced-learn pipeline
rf_pipeline = ImbPipeline([
    ('tfidf', tfidf_vectorizer),
    ('smote', smote),
    ('classifier', rf_classifier)
])

# Train the model using the pipeline
rf_pipeline.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf_pipeline.predict(X_test)

# Print a success message
print("Random Forest model successfully trained.")
```

### 6.2 Logistic Regression



```python
# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5500, ngram_range=(1,2))

# Define SMOTE
smote = SMOTE(random_state=42)

# Define the Logistic Regression classifier
lr_classifier = LogisticRegression(random_state=GLOBAL_CONSTANT, max_iter=1000)

# Create the pipeline
logistic_pipeline = ImbPipeline([
    ('tfidf', tfidf_vectorizer),
    ('smote', smote),
    ('classifier', lr_classifier)
])

# Train the model using the pipeline
logistic_pipeline.fit(X_train, y_train)

# Make predictions on the test set
y_pred_logistic_pipeline = logistic_pipeline.predict(X_test)

# Print a success message
print("Logistic Regression model successfully trained.")
```

### 6.3 Support Vector Machine



```python
# Create a pipeline for SVM with SMOTE
svm_pipeline = make_pipeline(
    TfidfVectorizer(max_features=5500, ngram_range=(1, 2)),
    SMOTE(random_state=GLOBAL_CONSTANT),
    SVC(kernel='rbf', gamma=1, C=10, random_state=42)
)

# Train the model using the pipeline
svm_pipeline.fit(X_train, y_train)

# Make predictions on the test set
y_pred_svm_pipeline = svm_pipeline.predict(X_test)

# Print a success message
print("Support Vector Machine model successfully trained.")
```

### 6.4 Extra Trees



```python
# Create a pipeline for Extra Trees with SMOTE
extra_trees_pipeline = make_pipeline(
    TfidfVectorizer(max_features=5500, ngram_range=(1, 2)),
    SMOTE(random_state=GLOBAL_CONSTANT),
    ExtraTreesClassifier(n_estimators=100, random_state=42)
)

# Train the model using the pipeline
extra_trees_pipeline.fit(X_train, y_train)

# Make predictions on the test set
y_pred_extra_trees = extra_trees_pipeline.predict(X_test)

# Print a success message
print("Extra Tress model successfully trained.")
```

### 6.5 Multi-Layer Perceptron (MLP) Classifier




```python
# Create a pipeline for MLPClassifier with SMOTE
mlp_pipeline = make_pipeline(
    TfidfVectorizer(max_features=5500,ngram_range=(1, 2)),
    SMOTE(random_state=GLOBAL_CONSTANT),
    MLPClassifier(random_state=GLOBAL_CONSTANT, max_iter=200)
)

# Train the MLPClassifier using the pipeline
mlp_pipeline.fit(X_train, y_train)

# Make predictions on the test set
y_pred_mlp = mlp_pipeline.predict(X_test)

# Print a success message
print("MLP model successfully trained.")
```

⚡ **Description: Model Evaluation** ⚡

Model Evaluation is a crucial stage in the machine learning process, where the performance and effectiveness of a trained model are assessed. This phase is particularly important for understanding how well the model performs on new, unseen data. Key aspects of Model Evaluation in the context of a sentiment classification model include:

- **Performance Metrics**: Utilizing various metrics such as accuracy, precision, recall, and F1-score to quantify the model's effectiveness.
- **Generalization Ability**: Analyzing how well the model generalizes to new, unseen tweets, beyond the data it was trained on.
- **Predictive Accuracy**: Assessing the model’s ability to accurately predict sentiment labels for tweets.

Through systematic evaluation, we gain insights into the model's strengths and weaknesses, guiding further improvements or adjustments.

---


The performance of each model will be evaluated based on the **precision**, **accuracy** and **F1 score** achieved.We will be looking at the following to determine and visualize these metrics:

 1. **Classification reports:**
 - **F1 score** - The F1 Score is a single metric that balances a model's ability  
 to correctly identify positive instances and its ability to avoid false positives, providing a combined measure of precision and recall.

 - **Accuracy** - Accuracy is a metric that assesses the overall correctness of a model's predictions by calculating the ratio of correctly predicted instances to the total number of instances.

 - **Precision** - Precision measures the accuracy of positive predictions made by a model, specifically the ratio of true positive predictions to the total number of positive predictions. It focuses on minimizing the rate of incorrect positive predictions.


 2. **Confusion Matrix:**
 - A table that summarizes the performance of a classification model by displaying the counts of true positive, true negative, false positive, and false negative predictions, allowing for the assessment of model accuracy and error types.

While our project endeavors to classify sentiments in climate change-related tweets and contributes valuable insights, it is important to acknowledge certain limitations that may impact the robustness and generalizability of our findings:

1. **Subjectivity in Sentiment Analysis:**
   - Sentiment analysis inherently involves subjective interpretation, and individuals may express sentiments in diverse ways, making it challenging to capture the full spectrum of opinions accurately.
   
2. **Evolution of Language:**
   - Language use is dynamic, and the evolution of terms and phrases over time can pose challenges. The model may not capture emerging trends or changes in language, potentially leading to outdated sentiment classifications.  

3. **Contextual Ambiguity:**
   - Tweets often contain context-specific nuances, humor, or sarcasm that may be challenging for the model to accurately interpret, potentially leading to misclassifications.   

4. **Hashtag Variability:**
   - While hashtags provide valuable insights, their meanings can evolve, and users may employ them inconsistently. The sentiment associated with a particular hashtag may vary among users, introducing ambiguity.  

5. **Geographical and Cultural Variations:**
   - Public sentiments on climate change can vary based on geographical and cultural factors. Our model may not fully account for these variations, limiting its generalizability to diverse populations.   

6. **Bias in Training Data:**
   - If the training data exhibits biases, the model may inadvertently learn and perpetuate those biases. Ensuring a diverse and representative training dataset is crucial to mitigate potential bias in sentiment classification.  

7. **Limited Scope of Sentiment Classes:**
   - Our chosen sentiment categories (Pro, Anti, Neutral, News) may not capture the full complexity of sentiments expressed in climate change discussions, potentially oversimplifying the rich tapestry of opinions.

8. **Dynamic Nature of Social Media:**
   - Social media platforms are dynamic, with rapidly changing trends and user behaviors. The model may struggle to adapt to the evolving nature of online discussions and sentiments.

9. **Privacy Concerns:**
   - Ensuring privacy compliance is crucial when working with user-generated content. Ethical considerations and privacy concerns may arise, especially when dealing with publicly available but sensitive information.
   
10. **Model Overfitting:**
    - Overfitting, where the model becomes too tailored to the training data, may limit its ability to generalize to new, unseen data. Regular model evaluation and potential adjustments are necessary to address overfitting issues.

Understanding these limitations is essential for interpreting the results of our sentiment analysis accurately. Ongoing refinement of the model, continual updates to training data, and consideration of ethical implications will contribute to addressing these challenges and enhancing the overall reliability of our machine learning application.
